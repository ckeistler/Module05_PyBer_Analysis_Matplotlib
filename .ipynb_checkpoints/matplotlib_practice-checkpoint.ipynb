```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Change the line color to green.
l1, = plt.plot(x_axis, y_axis, marker="*", color="b", linewidth=2, label='Boston')
l1.set_color('g')
# Change the marker to a diamond.
l1.set_marker('d')
# Change the line width to 2 points.
l1.set_linewidth(2)
# Add a legend for the city of Boston.
plt.legend("Boston")
# Add a title and axes labels.
plt.title("PyBer Far by Month")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Add a y-axis limit.
plt.ylim(0,45)
# Add grid lines.
plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x, y, 'o')
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x,y,color="r", label="Chicago")
plt.legend()
plt.ylabel("Fare($)")
plt.xlabel("Date")
plt.title("PyBer Fare by Month")
plt.ylim(0,50)
plt.gca().invert_xaxis()
```

```python
plt.scatter(y,x,color="r", label="Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x, y, s=y)
```

```python
y_axis_larger = []
for data in y:
  y_axis_larger.append(data*3)
plt.scatter(x, y, s=y_axis_larger)
```

```python
plt.scatter(x, y, s = [i * 3 for i in y])
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s=y)
```

```python
fig, ax = plt.subplots()
ax.scatter(y, x, s=[i * 5 for i in y], color='xkcd:sky blue', alpha=0.8, edgecolor="black", linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.pie(y, labels=x)
plt.show()
```

```python
# assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y, explode=explode_values, colors=colors, labels=x, autopct='%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y, explode=explode, colors=colors, shadow=True, startangle=96, counterclock=False, labels=x, autopct='%.1f%%')
fig.set_size_inches(8,8)
plt.show()
```
